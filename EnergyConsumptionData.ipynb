```python
#get data for each state and add to region
#north
for state in north:
    try:
        #get coal usage
        query_url_coal = f"{url}series_id=ELEC.CONS_TOT.COW-{state}-99.M&api_key={api_key}&out=json"
        response_coal = requests.get(query_url_coal).json()
        extract_coal = response_coal['series'][0]['data']
        #clean up nones
        corr_coal = []
        for m, i in extract_coal: #m = month, i - index
            if str(i) == 'None':
                corr_coal.append([m,0])
            else:
                corr_coal.append([m,float(i)])
    except:
        corr_coal = no_usage_list
    
    try:
        #get petroleum usage (includes petroleum coke after 2006)
        query_url_pet = f"{url}series_id=ELEC.CONS_TOT.PEL-{state}-99.M&api_key={api_key}&out=json"
        response_pet = requests.get(query_url_pet).json()
        extract_pet = response_pet['series'][0]['data']
        #clean up nones
        corr_pet = []
        for m, i in extract_pet: #m = month, i - index
            if str(i) == 'None':
                corr_pet.append([m,0])
            else:
                corr_pet.append([m,float(i)])
    except:
        corr_pet = no_usage_list
    
    try:
        #get natural gas usage
        query_url_natgas = f"{url}series_id=ELEC.CONS_TOT.NG-{state}-99.M&api_key={api_key}&out=json"
        response_natgas = requests.get(query_url_natgas).json()
        extract_natgas = response_natgas['series'][0]['data']
        #clean up nones
        corr_natgas = []
        for m, i in extract_natgas: #m = month, i - index
            if str(i) == 'None':
                corr_natgas.append([m,0])
            else:
                corr_natgas.append([m,float(i)])
    except:
        corr_natgas = no_usage_list
    
    #add the series to get totals
    data_sum = []
    i=0
    for pair in corr_coal:
        sumEnergy = corr_coal[i][1] + corr_pet[i][1] +corr_natgas[i][1]
        data_sum.append([corr_coal[i][0],sumEnergy])
        i = i+1
    data_north.append(data_sum)
    
#confirm output
print(f"States in North: {len(north)} States in Data: {len(data_north)}")
```
