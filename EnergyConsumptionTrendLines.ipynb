```python
#Import Dependencies
#pandas dataframes and statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#curve fitting
from scipy.stats import linregress
#CSV usage
import os
import csv
import pathlib
```

```python
#Load Region Data from CSV into dataframe
allRegions_df = pd.read_csv('All_Regions_Energy_Usage_By_Month.csv') 
allRegions_df.sort_values(by = 'Unnamed: 0', ascending = True, inplace = True)

#Grab months as an index list for scatterplots
months = allRegions_df.iloc[:, 0]

#Display Dataframe
allRegions_df
```

```python
#Create trend lines
#North
usage_North = allRegions_df['North']
plt.scatter(months,usage_North)

# Create equation of line
(slope, intercept, rvalue, pvalue, stderr) = linregress(months,usage_North)
regress_values_North = months * slope + intercept

#Find percentage increase (slope uses month as an integer)
rate = usage_North.pct_change().mean() * 100

#Plot the graph
plt.plot(months,regress_values_North,"r-")
plt.title('Energy Usage Trends for Region - North')
plt.xlabel('Month')
plt.ylabel('Energy Usage (MMBtu)')
plt.xticks(rotation=90)
print(f"The monthly rate of increase of Energy Usage for Region North : {round(rate,2)}%")
plt.show()
```

```python
#Plot all the trend lines for comparison
plt.plot(months,regress_values_North, "r-", label='North')
plt.plot(months,regress_values_South, "b-", label='South')
plt.plot(months,regress_values_Mid, "g-", label='Mid-West')
plt.plot(months,regress_values_West, "k-", label='West')
plt.title('Energy Consumption Trends Throughout the USA')
plt.xlabel('Month (Jan-2001 through Nov-2021)')
plt.ylabel('Energy Usage (MMBtu)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
