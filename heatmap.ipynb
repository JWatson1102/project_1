```python
latlng_csv_file = "state_lat_long.csv"
```

```python
lat_lng_df = pd.read_csv(latlng_csv_file)

latlng_df = lat_lng_df.drop(['City'], axis=1)
latlng_df = latlng_df.drop(labels=[1,8,11], axis=0)
latlng_df.head()
```

```python
# Configure gmaps with API key
gmaps.configure(api_key=gkey)
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = latlng_df[["Latitude", "Longitude"]].astype(float)

# Convert Poverty Rate to float and store
# HINT: be sure to handle NaN values

average_rate = state_avg_df["Avg State Price"].astype(float)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=average_rate, 
                                 dissipating=False, max_intensity=20,
                                 point_radius = 3,)

fig.add_layer(heat_layer)

fig
```
