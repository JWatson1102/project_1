```python
coal_df = pd.DataFrame()

for state in state_codes:
    url = f"http://api.eia.gov/series/?api_key={api_key}&series_id=ELEC.COST.COW-{state}-99.M"
    response = requests.get(url).json()
    try:
            
        response_data = response['series'][0]['data']
    
        month = []
        price = []
        
        for x in range(167):
            month.append(response_data[x][0])
            price.append(response_data[x][1])

        coal_df[f'{state} Price'] = price  

    except:
        price = []
        for x in range(167):
            price.append('N/A')

coal_df['Date'] = month

coal_df
```
