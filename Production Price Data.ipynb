```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census
from config import api_key
```

```python
state_codes = ['AL','AZ','AR','CA','CO','CT','DE','FL','GA','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY']
```

```python
coal_df = pd.DataFrame()

coal_check = True
date = []

for state in state_codes:
    url = f"https://api.eia.gov/series/?api_key={api_key}&series_id=ELEC.COST.COW-{state}-99.M"
    
    response = requests.get(url).json()
    try:
            
        response_data = response['series'][0]['data']
        
        price = []
        
        for x in range(167):
            price.append(response_data[x][1])
            if coal_check == True:
                date.append(response_data[x][0])
                
        coal_check = False
        coal_df[f'{state} Price'] = price  

    except:
        price = []
        for x in range(167):
            price.append(0)
        coal_df[f'{state} Price'] = price 
coal_df['Date (YYYYMM)'] = date
```

```python
coal_df = coal_df.fillna(0)

coal_df = coal_df.set_index('Date (YYYYMM)')
coal_df.sort_index(ascending=True, inplace=True)
coal_df
```

```python
petroleum_df = pd.DataFrame()

petroleum_check = True
date = []

for state in state_codes:
    url = f"https://api.eia.gov/series/?api_key={api_key}&series_id=ELEC.COST.PEL-{state}-99.M"

    response = requests.get(url).json()
    try:
            
        response_data = response['series'][0]['data']
        
        price = []
        
        for x in range(167):
            price.append(response_data[x][1])
            if petroleum_check == True:
                date.append(response_data[x][0])

        petroleum_check = False
        petroleum_df[f'{state} Price'] = price  

    except:
        price = []
        for x in range(167):
            price.append(0)
        petroleum_df[f'{state} Price'] = price 
petroleum_df['Date (YYYYMM)'] = date
```

```python
petroleum_df = petroleum_df.fillna(0)

petroleum_df = petroleum_df.set_index('Date (YYYYMM)')
petroleum_df.sort_index(ascending=True, inplace=True)
petroleum_df
```

```python
natural_gas_df = natural_gas_df.fillna(0)

natural_gas_df = natural_gas_df.set_index('Date (YYYYMM)')
natural_gas_df.sort_index(ascending=True, inplace=True)
natural_gas_df
```
