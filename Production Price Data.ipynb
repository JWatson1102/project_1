```python
dataframe = pd.DataFrame()

for state in state_codes:
    url = f"http://api.eia.gov/series/?api_key={api_key}&series_id=ELEC.SALES.{state}-ALL.M"
    response = requests.get(url).json()
    response_data = response['series'][0]['data']
    
    month = []
    price = []
    
    for x in range(251):
        month.append(response_data[x][0])
        price.append(response_data[x][1])
        
    dataframe[f'{state} Price'] = price  
dataframe['Date'] = month
```
